```python
import requests
from bs4 import BeautifulSoup as bs
import time
import re
import json
acount = 0
count = 0

playload = {
'from':'/bbs/Gossiping/index.html',
'yes':'yes'
}
rs = requests.session()
res = rs.post('https://www.ptt.cc/ask/over18',verify=False,data=playload)
page_url='https://www.ptt.cc/bbs/Gossiping/index{}.html'

for i in xrange(1816,15170):
    url = 'https://www.ptt.cc'
    res = rs.get(page_url.format(i),verify=False)
    soup = bs(res.text)
    div = soup.select('.title')
    #titlea = soup.select('.title a')
    #for s in xlen(titlea):
        #title = titlea[s].text
        #print title
    for page in div:
        bcount = 0
        ccount = 0
        gcount = 0
        try:
            numOfLike = page.select('a')[0].text
        except:
            numOfLike = '0'
        if((not numOfLike.isdigit()) & ('X' not in numOfLike)):
            url1 = page.select('a')[0]['href']
            #print url1
        pages_url = url+url1
        res1 = rs.get(pages_url,verify=False)
        soup1 = bs(res1.text)
        try:
            title = soup1.select('.article-meta-value')[2].text
        except IndexError: 
            acount+=1
            print 'IndexError:'+str(acount)
        #print title
        #title = title_class.text.split('] ')[1]
        #href = soup1.select('.f2 a')[0]['href']
        push = soup1.select('.push')
        message = ""
        for page in push:
            x = page.text
            message += x.encode('utf-8')
        push1 = soup1.select('.push-tag')
        for good in push1:
            if good.text == u'推 ':
                gcount += 1
                #print '推:'+str(acount)
            elif good.text == u'噓 ':
                bcount += 1
                #print '噓:'+str(bcount)
            else:
                ccount += 1
                #print '→:'+str(ccount)
            #print i.text
        #good = str(gcount)
        #bad = str(bcount)
        #no = str(ccount)
        #print message
        try:
            date_col = soup1.select('.article-meta-value')[3]
            month = date_col.text.split(' ')[1]
            if date_col.text.split(' ')[1] == "Jan":
                y = re.sub('Jan','01',month)
            elif date_col.text.split(' ')[1] == "Feb":
                y = re.sub('Feb','02',month)
            elif date_col.text.split(' ')[1] == "Mar":
                y = re.sub('Mar','03',month)
            elif date_col.text.split(' ')[1] == "Apr":
                y = re.sub('Apr','04',month)
            elif date_col.text.split(' ')[1] == "May":
                y = re.sub('May','05',month)
            elif date_col.text.split(' ')[1] == "Jun":
                y = re.sub('Jun','06',month)
            elif date_col.text.split(' ')[1] == "Jul":
                y = re.sub('Jul','07',month)
            elif date_col.text.split(' ')[1] == "Aug":
                y = re.sub('Aug','08',month)
            elif date_col.text.split(' ')[1] == "Sep":
                y = re.sub('Sep','09',month)
            elif date_col.text.split(' ')[1] == "Oct":
                y = re.sub('Oct','10',month)
            elif date_col.text.split(' ')[1] == "Nov":
                y = re.sub('Nov','11',month)
            elif date_col.text.split(' ')[1] == "Dec":
                y = re.sub('Dec','12',month)
            if len(date_col.text.split(' '))<=5:
                date = date_col.text.split(' ')[4],y,date_col.text.split(' ')[2]
            else:
                date = date_col.text.split(' ')[5],y,'0'+date_col.text.split(' ')[3]
            date_new = date[0]+date[1]+date[2]
        except IndexError:
            print pages_url
        try:
            main_content= soup1.select('#main-content')[0]
        except IndexError:
            print 'content'+pages_url
        for l in main_content.select('span'):
            main_content.span.decompose()

        dic ={
            'comp':'ptt',
            'url':pages_url,
            'title':title,
            'date':date_new,
            'content':main_content.text,
            'message':message,
            'good':gcount,
            'bad':bcount,
            'no':ccount
        }
        data = json.dumps(dic)
        name = date_new+'_'+pages_url.split('/')[5].split('.')[1]
        try:
            f = open('E:\crawler\{}.json'.format(name),'w')
            f.write(data)
            f.close()
        except IOError:
            print pages_url
        time.sleep(0.1)
```

### PTT Beautiful

```python
import requests
import shutil
import random
from time import sleep
from urlparse import urljoin
from bs4 import BeautifulSoup
count = 0
xcount = 0
ycount = 0
f_url = 'https://www.ptt.cc/bbs/Beauty/index{}.html'
for i in xrange(1740,1750):
    url=f_url.format(i) 
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    div = soup.select('.title')
    for title in div :
        try:
            numOfLike = title.select('a')[0].text
        except:
            numOfLike = '0'
        if((not numOfLike.isdigit()) & ('X' not in numOfLike)):
            x = title.select('a')[0]['href']
            res1 = urljoin(url, x) 
            res1 = requests.get(urljoin(url,x))
            soup1 = BeautifulSoup(res1.text)
            div1 = soup1.select('.richcontent')
            for pic in div1:
                try:
                    y = pic.select('img')[0]['src']
                except IndexError:
                    count+=1
                    print 'IndexError:'+str(count)            
                try: 
                    if(y[0:4]=='http'):
                        imgurl = requests.get(y, stream=True)
                    else:
                        imgurl = requests.get('http:'+ y, stream=True)       
                except:
                    xcount+=1
                    print 'InvalidSchema:'+str(xcount)
                name = y.split('/')[-1]
                try:
                    f = open('E:\img\{}'.format(name),'wb')
                except:
                    ycount+=1
                    print 'IOError:'+str(ycount)
                imgurl.raw.decode_content = True
                shutil.copyfileobj(imgurl.raw, f)
                f.close()
                sleep(round(random.uniform(0,2),3))
```
